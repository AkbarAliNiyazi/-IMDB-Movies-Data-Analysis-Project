# 20. Classify Movies Based on Rating [Excellent, Good and Average]

```python
df.columns
```

```python
def rating(rating):
    if rating >= 7.0:
        return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else:
        return "Average"
```

# 22. Find Unique Values From Genre

```python
df.columns
```

```python
df['Genre']
```

```python
list = []
for value in df['Genre']:
    list.append(value.split(','))
```

```python
list
```

```python
one_d = []
for item in list:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list = []
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
